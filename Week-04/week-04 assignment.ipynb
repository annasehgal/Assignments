We will build a complex plot about the poverty rate using the `Crime Data` dataset.  Complex plots are often built up, little by little, as we will do below.

1. Using the `Crime Data` dataset, create a density plot of the distribution of `poverty` rate (x) by `region` (fill). To avoid problems with overplotting, use `alpha=.5` inside the density geometry to make the colors transparent.

```python
import pandas as pd
import plotnine as pn
```

```python
p = (pn.ggplot(data = crime_data_df)
    + pn.geom_density(pn.aes(x = "poverty", fill = "region"), alpha = 0.5))
p.draw()
```

2. These are pretty diverse regions. Let's see how the poverty rate varies within region. Fill by division instead, and create sub-plots by region. To create sub-plots, add the layer `pn.facet_wrap("region")`.

```python
p += pn.facet_wrap("region")
p.draw()
```

3. Add a different theme to this plot. _Hint: Do not use theme_void() for this step_

```python
p += pn.theme_minimal()
p.draw()
```

4. Add the layer `pn.scale_fill_brewer()` to change the color palette.

```python
p += pn.scale_fill_brewer()
p.draw()
```

5. Capitalize the legend title. _Hint, you're adding something to the_ `scale_` _section. Be sure that you continue to build off your code from the previous steps so that your theme is still present_.

```python
p += pn.guides(fill=pn.guide_legend(title="Region"))
p.draw()
```

6. Let's explore another plot of the same data.  We'll next consider boxplots instead of density plots.  Start by setting up vertical boxplots, putting your continuous variable on the y-axis.

```python
p2 = (pn.ggplot(data = crime_data_df) +
     pn.geom_boxplot(pn.aes(x='division', y='poverty')))
p2.draw()
```

```python
#p2 = (pn.ggplot(data = crime_data_df) +
     # pn.geom_boxplot(pn.aes(x="division", y="poverty"), width=0.7)
     # + pn.theme(axis_text_x=pn.element_text(angle= 50, hjust=1))) #https://stackoverflow.com/questions/54550444/rotating-x-axis-labels-and-changing-theme-in-ggplot2
#p2.draw()
```

7. The x-axis is using numbers instead of names now. Let's fix that by clearly defining that x-axis is division. _Note, The division levels will appear to blend together, but we'll be fixing that in later steps._

```python
p2 += pn.scale_x_discrete(name="Division")
p2.draw()
```

8. In this case, division is nested within region and so we don't need ALL the levels of division to show up in each panel. Let's "Free" our x-axis by adding `scales="free_x"` to the `facet_wrap`.

```python
p2 += pn.facet_wrap("division", scales="free_x")
p2.draw()
```

9. Let's try these as horizontal plots next. Add a `coord_flip()` layer. This also helps with readability of long names. _Note, you will also need to change your scales to free the y-axis instead now._

```python
p2 += pn.coord_flip()
p2 += pn.facet_wrap("division", scales="free_y")
p2.draw()
```

10. Remove the legend by adding `guide='none'` to the `scale` layer.

```python
p2 += pn.theme(legend_position="none")
p2.draw()
```

11. Add `ncol=1` to the `facet_wrap` to stack the panels vertically.

```python
p2 += pn.facet_wrap("region", ncol=1)
p2.draw()
```

12. Add an appropriate title and axis titles.  _NOTE: Be sure your axes are correct, due to the effects of coord_flip()._

13. Save the figure as a .pdf file.  Set the keyword arguments `path`, `width`, and `height`.  The argument `path` accepts a file name with extension, such as "name_of_file.pdf".  The arguemnts `width` and `height` accept some numbers in inches, to specify the size of the saved figure.

```python
p2.save("poverty_rate_plot.pdf", width=9, height=10)
```
