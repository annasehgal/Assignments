6. What percent of records are missing data on weight gained during pregnancy? There are several ways you can calculate this. Explain what your process, and answer the question in a full sentence.

```python
missing_percent = (ncbirths["gained"].isna().sum() / ncbirths.shape[0]) * 100 # EAR .sum() Nice
# EAR nothing wrong with the parentheses you added, but 27 / 1000 * 100 will evaluate just the same,
# since Python treats both / and * as the same precedence and works left 27/1000 to right (27/1000) * 100
print("Percent of records missing weight gained data:", missing_percent)
```

## Police Shootings dataset

10. Execute the cell below to download and import the dataset Police Shootings from Dr. Donatello's webpage.  Read about the dataset on the webpage where the Washington Post's hosts the dataset: <https://github.com/washingtonpost/data-police-shootings>.

```python
police_shootings = pd.read_excel("https://www.norcalbiostat.com/data/fatal-police-shootings-data.xlsx")
```

11. Create (a new variable that holds) a boolean Series that is True whenever `race` is NA *or* `race` is equal to the letter "O".

```python
race_na_o = (police_shootings["race"].isna()) | (police_shootings["race"] == "O")
```

```python
race_na_o
```

12. Set the values of `race`, where `race` is NA or `race` is "O", equal to the string "unknown".

```python
police_shootings.loc[race_na_o, "race"] = "unknown"
```

```python
police_shootings
```

13. Change the type of the column `race` to type `category`.

```python
police_shootings["race"] = police_shootings["race"].astype("category")
police_shootings["race"]
```

14. Rename the categories of `race` according as follows:

* "A" -> "asian"
* "B" -> "black"
* "H" -> "hispanic"
* "N" -> "native"
* "W" -> "white"

```python
race_mapping = {"A": "asian", "B": "black", "H": "hispanic", "N": "native", "W": "white"}
police_shootings["race"] = police_shootings["race"].cat.rename_categories(race_mapping)
```

```python
police_shootings["race"]
```

15. Create a table for the counts of the values in the variable `race`.

```python
race_counts = police_shootings["race"].value_counts()
```

```python
race_counts
```

16. Calculate the proportions of the `race`s, relative to the total number of rows in `police_shootings`, that were shot by police in 2015.

```python
total_shootings_2015 = police_shootings[police_shootings["date"].dt.year == 2015].shape[0]
```

```python
race_proportions_2015 = race_counts / total_shootings_2015
```

```python
print(race_counts)
```

```python
print(race_proportions_2015)
```
